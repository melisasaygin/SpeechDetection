```python
import pandas as pd
from datetime import datetime
import os

# list of task numbers we are interested in
tasks_to_include = [12, 14, 16, 18, 20, 23, 25, 28, 30, 34, 38, 42, 
                    36, 40, 44, 47, 51, 55, 49, 53, 57, 60, 64, 68, 
                    62, 66, 70, 75, 78, 83, 87, 91, 85, 89, 93]

# folder path containing the timelabels files
timelabels_folder_path = r"C:\Users\msa583\Desktop\Re-train short timescale Acc model"

# dictionary to store start times and task labels for each participant
start_times_with_labels = {}

# function to parse date, keeping only the time part and using today's date
def safe_strptime(value, date_format='%d-%m-%y/%H:%M:%S'):
    try:
        if isinstance(value, str):  # Ensure the value is a string before parsing
            # Parse the date and time
            original_datetime = datetime.strptime(value, date_format)
            # Use today's date while keeping the time portion from original
            today = datetime.today().date()  # Get today's date
            return datetime.combine(today, original_datetime.time())  # Combine today's date with original time
        else:
            return pd.NaT  # returns Not a Time (NaT) for invalid formats
    except Exception as e:
        print(f"Error parsing date: {e}")
        return pd.NaT  # returns Not a Time if there's an error in parsing

# loop over each participant in your processed data
for participant_id in processed_data.keys():
    # Construct the filename for the task label file
    timelabels_filename = f"timelabelsPPID_{participant_id}.txt"
    timelabels_path = os.path.join(timelabels_folder_path, timelabels_filename)

    try:
        # read the task labels file (from tab separated text file)
        task_labels = pd.read_csv(
            timelabels_path,
            sep="\t",
            header=0,
            names=["StartTime", "StopTime", "Label"],
            dtype={"StartTime": str, "StopTime": str}  # Force StartTime and StopTime to be read as strings
        )

        # apply the safe_strptime function 
        task_labels["StartTime"] = task_labels["StartTime"].apply(safe_strptime)

        # filter rows for the tasks of interest
        filtered_tasks = task_labels[task_labels["Label"].isin(tasks_to_include)]

        # create a dictionary for the participant with task labels as keys and start times as values
        start_times_with_labels[participant_id] = {
            label: start_time for label, start_time in zip(filtered_tasks["Label"], filtered_tasks["StartTime"])
        }

        # print message to track progress
        print(f"Stored start times for participant {participant_id}: {start_times_with_labels[participant_id]}")

    except FileNotFoundError:
        print(f"Task label file for participant {participant_id} not found.")
    except Exception as e:
        print(f"Error processing task label file for participant {participant_id}: {e}")

# print the start times and labels for all participants
print(start_times_with_labels)
```

```python
durations = {
    12: 180, 14: 180, 16: 180, 18: 90, 
    20: 90, 23: 90, 25: 90, 28: 90, 30: 90, 
    34: 30, 38: 30, 42: 30, 36: 30, 40: 30, 44: 30,
    47: 30, 51: 30, 55: 30, 49: 30, 53: 30, 57: 30,
    60: 30, 64: 30, 68: 30, 62: 30, 66: 30, 70: 30,
    75: 90, 78: 90, 83: 30, 87: 30, 91: 30, 85: 30, 89: 30, 93: 30
}  
import pandas as pd

for participant_id, df in processed_data.items():
    # initialize new column with NaN or a default value
    df["Task_Label"] = None

    # retrieve start times for the present participant
    participant_start_times = start_times_with_labels.get(participant_id, {})

    for task_label, start_time in participant_start_times.items():
        # check if there is a duration for this task
        duration = durations.get(task_label, None)
        if duration is None:
            continue  # skip the task number if no duration is specified for the task label

        # calculate the end time for this task
        end_time = start_time + pd.Timedelta(seconds=duration)

        # assign the task label for the duration specified
        mask = (df.index >= start_time) & (df.index < end_time)
        df.loc[mask, "Task_Label"] = task_label

# after we run this, each processed_data[participant_id] will have a "Task_Label" column with labels based on start times and durations
```

```python
import numpy as np
import pandas as pd
import librosa
from scipy.signal import butter, filtfilt
from scipy.stats import kurtosis, skew, entropy

def butterworth_highpass(data, cutoff_frequency=75, sampling_rate=1000, order=2):
    b, a = butter(order, cutoff_frequency, btype='high', fs=sampling_rate, analog=False)
    y = filtfilt(b, a, data)
    return y

tasks = [12, 14, 16, 18, 20, 23, 25, 28, 30, 34, 36, 38, 40, 42, 44, 47, 49, 51, 53, 55, 57, 
         60, 62, 64, 66, 68, 70, 75, 78, 83, 85, 87, 89, 91, 93]

# initialize a list to store features for each task
all_features = []

# loop over each participant in processed_data
for participant_id, df_acc in processed_data.items():
    task_segments_count = {task: 0 for task in tasks}  # Track number of segments per task

    for task in tasks:
        # find all starting indices of the task
        task_indices = df_acc[df_acc['Task_Label'] == task].index

        if len(task_indices) > 0:
            # get the starting position for indexing with `.iloc`
            start_pos = df_acc.index.get_loc(task_indices[0])
            task_duration_samples = df_acc[df_acc['Task_Label'] == task].shape[0]
            num_segments = task_duration_samples // 3000  # Each segment is 3 seconds (3000 samples)

            # update the task segment count
            task_segments_count[task] = num_segments

            for i in range(num_segments):
                # extract the 3000-sample (3 seconds) segment using .iloc
                segment_start = start_pos + i * 3000
                acc_z = df_acc.iloc[segment_start:segment_start + 3000]['MZR']
                y_prefilter = acc_z.to_numpy()
                y = butterworth_highpass(y_prefilter)
                y_demeaned = y - np.mean(y)

                # calculate spectral features for the 3 seconds segment
                sr = 1000
                zcr = np.mean(librosa.feature.zero_crossing_rate(y=y_demeaned))
                spectral_centroid = np.mean(librosa.feature.spectral_centroid(y=y, sr=sr, n_fft=2048, hop_length=512))
                spectral_bandwidth = np.mean(librosa.feature.spectral_bandwidth(y=y, sr=sr, n_fft=2048, hop_length=512))
                spectral_flatness = np.mean(librosa.feature.spectral_flatness(y=y, n_fft=2048, hop_length=512))
                spectral_rolloff_25 = np.mean(librosa.feature.spectral_rolloff(y=y, sr=sr, n_fft=2048, hop_length=512, roll_percent=0.25))
                spectral_rolloff_50 = np.mean(librosa.feature.spectral_rolloff(y=y, sr=sr, n_fft=2048, hop_length=512, roll_percent=0.50))
                spectral_rolloff_85 = np.mean(librosa.feature.spectral_rolloff(y=y, sr=sr, n_fft=2048, hop_length=512, roll_percent=0.85))
                rms = np.mean(librosa.feature.rms(y=y, frame_length=2048, hop_length=512))

                spectrum = np.fft.rfft(y)
                spectral_kurtosis = kurtosis(abs(spectrum))
                spectral_skewness = skew(abs(spectrum))
                spectral_entropy = entropy(abs(spectrum), base=2)
                spectral_variance = np.var(abs(spectrum))
                spectral_mean = np.sum(abs(spectrum)) / len(spectrum)

                def crest(spectrum):
                    absSpectrum = abs(spectrum)
                    spectralSum = np.sum(absSpectrum)
                    maxFrequencyIndex = np.argmax(absSpectrum)
                    maxSpectrum = absSpectrum[maxFrequencyIndex]
                    return maxSpectrum / spectralSum

                crest_value = crest(spectrum)

                # append features for the current segment to the list
                features = [zcr, spectral_centroid, spectral_bandwidth, spectral_flatness,
                            spectral_rolloff_25, spectral_rolloff_50, spectral_rolloff_85, rms,
                            spectral_kurtosis, spectral_skewness, spectral_entropy,
                            spectral_variance, spectral_mean, crest_value, task, participant_id]
                all_features.append(features)
    
    # print the number of segments for each task after processing the present participant
    print(f"Participant {participant_id}:")
    for task in tasks:
        print(f"  Task {task}: {task_segments_count[task]} segments")

# at the end, create dataFrame from the features
feature_names = ['AccZ_ZCR', 'AccZ_Spectral_Centroid', 'AccZ_Spectral_Bandwidth', 'AccZ_Spectral_Flatness',
                 'AccZ_Spectral_Rolloff_25', 'AccZ_Spectral_Rolloff_50', 'AccZ_Spectral_Rolloff_85', 'AccZ_RMS', 
                 'AccZ_Spectral_Kurtosis', 'AccZ_Spectral_Skewness', 'AccZ_Spectral_Entropy', 
                 'AccZ_Spectral_Variance', 'AccZ_Spectral_Mean', 'AccZ_Spectral_Crest', 'Task_Label', 'Participant_ID']
df_features_acc_z = pd.DataFrame(all_features, columns=feature_names)

# display the dataframe
df_features_acc_z
```
